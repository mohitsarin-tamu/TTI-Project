```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv("a240402k.csv")

# Extract the features (X, Y, Z values)
X_values = df['X (in)']
Y_values = df['Y (in)']
Z_values = df['Z (in)']

# Combine X, Y, Z values into a single array
features = df[['X (in)', 'Y (in)', 'Z (in)']]

# Calculate the sum of squared distances for different values of k
sse = []
for k in range(1, 30):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    sse.append(kmeans.inertia_)



```

```python
# Plot the elbow curve
plt.plot(range(1, 30), sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances (SSE)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Load the CSV data
df = pd.read_csv("a240402k.csv")

# Extract the features (X, Y, Z values)
X_values = df['X (in)']
Y_values = df['Y (in)']
Z_values = df['Z (in)']

# Combine X, Y, Z values into a single array
features = df[['X (in)', 'Y (in)', 'Z (in)']]

# Determine the optimal number of clusters using the elbow method
# In this example, let's say we choose 3 clusters
num_clusters = 30

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(features)

# Initialize arrays to store maximum and minimum heights for each cluster
max_heights = []
min_heights = []

# Find maximum and minimum height within each cluster
for cluster_id in range(num_clusters):
    cluster_points = features[clusters == cluster_id]
    max_height = cluster_points['Z (in)'].max()
    min_height = cluster_points['Z (in)'].min()
    max_heights.append(max_height)
    min_heights.append(min_height)
    print(f"Cluster {cluster_id + 1}: Max Height = {max_height}, Min Height = {min_height}")

# Plot the clusters in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot each cluster with a different color
for cluster_id in range(num_clusters):
    cluster_points = features[clusters == cluster_id]
    ax.scatter(cluster_points['X (in)'], cluster_points['Y (in)'], cluster_points['Z (in)'], label=f'Cluster {cluster_id + 1}')

# Set legend position to the right-hand side
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Set labels and title
ax.set_xlabel('X (in)')
ax.set_ylabel('Y (in)')
ax.set_zlabel('Z (in)')
ax.set_title('Clusters in 3D')

# Add legend
ax.legend()

plt.show()
```

```python
# import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv("a240402b.csv")

# Extract the features (X, Y, Z values)
features = df[['X (in)', 'Y (in)', 'Z (in)']]

# Calculate the silhouette score for different values of k
silhouette_scores = []
for k in range(2, 30):  # Assuming you want to try k from 2 to 100
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(features)
    silhouette_avg = silhouette_score(features, cluster_labels)
    silhouette_scores.append(silhouette_avg)

    # getting less than 2
# Plot silhouette scores for different values of k
plt.plot(range(2, 30), silhouette_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Values of k')
plt.show()

# Find the best k based on the maximum silhouette score
best_k = silhouette_scores.index(max(silhouette_scores)) + 2  # Add 2 because k starts from 2
print(f"The best value of k is: {best_k}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
import numpy as np

# Load the CSV data
df = pd.read_csv("a240402k.csv")

# Extract the features (X, Y, Z values)
X_values = df['X (in)']
Y_values = df['Y (in)']
Z_values = df['Z (in)']

# Combine X, Y, Z values into a single array
features = df[['X (in)', 'Y (in)', 'Z (in)']]

# Determine the optimal number of clusters using the elbow method
# In this example, let's say we choose 3 clusters
num_clusters = 30

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(features)

# Initialize arrays to store maximum and minimum heights for each cluster
max_heights = []
min_heights = []

# Find maximum and minimum height within each cluster
for cluster_id in range(num_clusters):
    cluster_points = features[clusters == cluster_id]
    max_height = cluster_points['Z (in)'].max()
    center_point = cluster_points[cluster_points['Z (in)'] == max_height].iloc[0]
    radius = 0.5
    min_height = cluster_points[(cluster_points['X (in)'] - center_point['X (in)'])**2 + 
                                 (cluster_points['Y (in)'] - center_point['Y (in)'])**2 < radius**2]['Z (in)'].min()
    max_heights.append(max_height)
    min_heights.append(min_height)
    print(f"Cluster {cluster_id + 1}: Max Height = {max_height}, Min Height = {min_height}")

# Plot the histogram of height from the center of the maximum height
plt.hist(min_heights, bins=10, edgecolor='black')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Histogram of Minimum Height from Center of Maximum Height')
plt.show()

# Plot the clusters in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot each cluster with a different color
for cluster_id in range(num_clusters):
    cluster_points = features[clusters == cluster_id]
    ax.scatter(cluster_points['X (in)'], cluster_points['Y (in)'], cluster_points['Z (in)'], label=f'Cluster {cluster_id + 1}')

# Set labels and title
ax.set_xlabel('X (in)')
ax.set_ylabel('Y (in)')
ax.set_zlabel('Z (in)')
ax.set_title('Clusters in 3D')

# Add legend
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Load the CSV data
df = pd.read_csv("a240402b.csv")

# Extract the features (X, Y, Z values)
features = df[['X (in)', 'Y (in)', 'Z (in)']]

# Determine the optimal number of clusters using the elbow method
num_clusters = 30

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(features)

# Initialize lists to store max and min heights for each cluster
max_heights = []
min_heights = []

# Iterate over each cluster and plot individual graphs
for cluster_id in range(num_clusters):
    # Select points belonging to the current cluster
    cluster_points = features[clusters == cluster_id]
    
    # Calculate maximum height within the cluster
    max_height = cluster_points['Z (in)'].max()
    max_heights.append(max_height)

    # Calculate distances between the max height point and its neighboring points
    distances = np.sqrt((cluster_points['X (in)'] - cluster_points['X (in)'].mean())**2 + 
                        (cluster_points['Y (in)'] - cluster_points['Y (in)'].mean())**2)

    # Compute the percentiles of distances relative to the radius
    percentiles = np.linspace(50, 95, 46)  # From 50th to 95th percentile
    percentile_distances = np.percentile(distances, percentiles)

    # Calculate the heights of neighboring points corresponding to the percentiles of distances
    heights = []
    for percentile_distance in percentile_distances:
        height = cluster_points[distances <= percentile_distance]['Z (in)'].min()
        heights.append(height)
    
    # Calculate the 90th percentile of the x-axis values
    percentile_90th = np.percentile(percentiles, 90)
    # Find the corresponding height
    min_height = heights[int(percentile_90th) - 50]
    min_heights.append(min_height)

    # Print the maximum and minimum height for the current cluster
    print(f"Cluster {cluster_id + 1}: Max Height = {max_height}, Min Height = {min_height}, Range = {max_height - min_height}")

    # Plot the line graph for the current cluster
    plt.plot(percentiles, heights, marker='o')
    plt.xlabel('Percentile of Distance from Neighbor (50-95%)')
    plt.ylabel('Height of Neighboring Points')
    plt.title(f'Line Graph of Height vs Percentile of Distance from Neighbor (Cluster {cluster_id + 1})')
    plt.grid(True)
    plt.show()


    
```
